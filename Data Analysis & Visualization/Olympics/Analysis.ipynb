```python
import pandas as pd
import pyecharts.charts as pye
import pyecharts.options as opts
```

```python
df = pd.read_csv('athlete_events.csv')

df
```

```python
plot = (pye.WordCloud()
        .add(series_name='', 
             word_size_range=[10, 75],
             data_pair=[(k, v) for k, v in df['Sport'].value_counts().to_dict().items()]
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="Word Cloud of All Olympics sport programs", 
            ),
        )
        )

plot.render_notebook()
```

```python
data = df.groupby(['Year']).mean().reset_index()[['Year', 'Age']]
plot = (pye.Line()
        .add_xaxis(data['Year'].astype(str).to_list())
        .add_yaxis('', data['Age'].astype(int).to_list())
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Average Age Over Year"),
            xaxis_opts=opts.AxisOpts(name="Year"),
            yaxis_opts=opts.AxisOpts(
                name="Average Age",
                splitline_opts=opts.SplitLineOpts(is_show=True),
                is_scale=True,
            ),
        )
        )

plot.render_notebook()
```
