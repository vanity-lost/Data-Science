```python
import pandas as pd
import pyecharts.charts as pye
import pyecharts.options as opts
from IPython.display import HTML
```

```python
df = pd.read_csv('athlete_events.csv')

df
```

```python
plot = (pye.WordCloud()
        .add(series_name='', 
             word_size_range=[10, 75],
             data_pair=[(k, v) for k, v in df['Sport'].value_counts().to_dict().items()]
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="Word Cloud of All Olympics sport programs", 
            ),
        )
        )

plot.render_notebook()
```

```python
plot.render('Word Cloud of All Olympics sport programs.html')
```

```python
data = df.groupby(['Sex', 'Year']).mean().reset_index()[['Sex', 'Year', 'Age']].sort_values(by='Year')
plot = (pye.Line()
        .add_xaxis(list(data['Year'].astype(str).unique()))
        .add_yaxis('Female Age', data.loc[data['Sex']=='F', 'Age'].astype(int).to_list(), is_smooth=True)
        .add_yaxis('Male Age', data.loc[data['Sex']=='M', 'Age'].astype(int).to_list(), is_smooth=True)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Average Age Over Year"),
            xaxis_opts=opts.AxisOpts(name="Year"),
            yaxis_opts=opts.AxisOpts(
                name="Average Age",
                splitline_opts=opts.SplitLineOpts(is_show=True),
                is_scale=True,
            ),
        )
        )

plot.render_notebook()
```

```python
plot.render('Average Age Over Year.html')
```

```python
data = df.loc[df['Medal']=='Gold', ['Team', 'Medal']].groupby(['Team']).count().reset_index().sort_values(by='Medal')[-10:]
plot = (pye.Bar()
        .add_xaxis(data['Team'].to_list())
        .add_yaxis('Gold Medal', data['Medal'].to_list())
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Top 10 Teams with most gold medals"),
            xaxis_opts=opts.AxisOpts(name="Team", axislabel_opts=opts.LabelOpts(rotate=-15)),
            yaxis_opts=opts.AxisOpts(
                name="Total Gold Medal",
                splitline_opts=opts.SplitLineOpts(is_show=True),
                is_scale=True,
            ),
        )
        )

plot.render_notebook()
```

```python
plot.render('Top 10 Teams with most gold medals.html')
```

```python
data = (df.loc[(df['Team']=='China') & (df['Medal']=='Gold'), ['Sport', 'Medal']]
        .groupby(['Sport']).count().reset_index()
        .sort_values(by='Medal')[-10:])
plot = (pye.Funnel(init_opts=opts.InitOpts(width="1000px"))
        .add('China Gold Medal', [[i.Sport, i.Medal] for i in data.itertuples()],
             label_opts=opts.LabelOpts(is_show=True, position="inside"),)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="China's Top 10 Sport with most gold medals"),
            legend_opts=opts.LegendOpts(type_="scroll", pos_top='30%', pos_left="right", orient="vertical"),
        )
        )

plot.render_notebook()
```

```python
plot.render('China\'s Top 10 Sport with most gold medals.html')
```
