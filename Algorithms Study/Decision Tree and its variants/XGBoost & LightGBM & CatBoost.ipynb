# XGBoost & LightGBM & CatBoost

```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier
```

```python
iris = load_iris()
```

# XGBoost

```python
xgb_clf = xgb.XGBClassifier(use_label_encoder=False)
xgb_clf.fit(iris.data, iris.target)
y_pred = xgb_clf.predict(iris.data)
print('AUC of testset based on XGBoost: ', accuracy_score(y_pred, iris.target))
```

# LightGBM

```python
lgb_clf = lgb.LGBMClassifier()
lgb_clf.fit(iris.data, iris.target)
y_pred = lgb_clf.predict(iris.data)
print('AUC of testset based on LightGBM: ', accuracy_score(y_pred, iris.target))
```

# CatBoost

```python
cat_clf = CatBoostClassifier(iterations=100)
cat_clf.fit(iris.data, iris.target, verbose=False)
y_pred = cat_clf.predict(iris.data)
print('AUC of testset based on CatBoost: ', accuracy_score(y_pred, iris.target))
```
