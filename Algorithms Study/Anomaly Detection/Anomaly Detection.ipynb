# Isolation Forest & Local Outlier Factor & SVMs

```python
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = list(map(lambda x: color_mapping[x], y))
```

```python
model = PCA(n_components=2, random_state=0)
model = model.fit_transform(X)
plt.figure(figsize=(10, 6))
plt.scatter(model[:, 0], model[:, 1], c=colors)
plt.show()
```

## Isolation Forest

```python
iso = IsolationForest(contamination=0.1)
yhat = iso.fit_predict(X)
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = list(map(lambda x: color_mapping[x], y))
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = [colors[i] if yhat[i]==1 else sns.xkcd_rgb['red'] for i in range(len(colors))]
plt.figure(figsize=(10, 6))
plt.scatter(model[:, 0], model[:, 1], c=colors)
plt.show()
```

## Local Outlier Factor

```python
lof = LocalOutlierFactor()
yhat = lof.fit_predict(X)
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = list(map(lambda x: color_mapping[x], y))
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = [colors[i] if yhat[i]==1 else sns.xkcd_rgb['red'] for i in range(len(colors))]
plt.figure(figsize=(10, 6))
plt.scatter(model[:, 0], model[:, 1], c=colors)
plt.show()
```

```python
ee = OneClassSVM(nu=0.01)
yhat = ee.fit_predict(X)
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = list(map(lambda x: color_mapping[x], y))
color_mapping = {0: sns.xkcd_rgb['bright purple'], 1: sns.xkcd_rgb['lime'], 2: sns.xkcd_rgb['ochre']}
colors = [colors[i] if yhat[i]==1 else sns.xkcd_rgb['red'] for i in range(len(colors))]
plt.figure(figsize=(10, 6))
plt.scatter(model[:, 0], model[:, 1], c=colors)
plt.show()
```
